## 1. Group by

The groupby() method allows you to group your data and execute functions on these groups.

```python
import pandas as pd
import numpy as np
```

```python
# Sample sales data
df = pd.DataFrame({
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Home'],
    'Sales': [200, 150, 100, 80, 300],
    'Profit': [50, 40, 20, 15, 70]
})

# Group by Category and sum the numeric columns
grouped_sum = df.groupby('Category').sum()
grouped_sum
```

```python
# groupby multiple columns

df.groupby(['Category', 'Sales']).mean()
```

```python
# Sum of sales and average of profit
df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'mean'})
```

```python
# keep index

df.groupby('Category', as_index=False).sum()
```

## 2. Transform (.transform)

Use this when you want to perform a group-based calculation but keep the result the same length as your original DataFrame. It’s perfect for normalizing data.

```python
df = pd.DataFrame({
    'Team': ['A', 'A', 'B', 'B', 'C'],
    'Score': [10, 20, 40, 50, 100]
})

# Calculate the group mean and "broadcast" it to every row
df['Team_Avg'] = df.groupby('Team')['Score'].transform('mean')

# calculate how much each row deviates from its team average
df['Deviation'] = df['Score'] - df['Team_Avg']
df
```

## 3. Filter (.filter)

This allows you to drop entire groups based on a condition. For example, if you only want to see data from teams that have a high average score.

```python
# Keep only teams where the average score is greater than 20
# Team A (avg 15) will be removed; Team B and C will stay.
filtered_df = df.groupby('Team').filter(lambda x: x['Score'].mean() > 20)
filtered_df
```

## 4. Apply (.apply)

For complex logic

```python
df = pd.DataFrame({
    'Team': ['A', 'A', 'B', 'B', 'C'],
    'Score': [10, 20, 40, 50, 100]
})

# Get the top 1 scorer for each team
def get_top_scorer(group):
    return group.nlargest(1, 'Score')

top_scorers = df.groupby('Team').apply(get_top_scorer)
top_scorers
```
