## loc vs iloc

.loc (Label-based): You use the names (labels) of rows and columns.

.iloc (Integer-based): You use the position (0, 1, 2...)—just like a standard Python list.

### 1. Using .loc (Label-based)

Syntax: df.loc[row_label, column_label]

```python
df = pd.DataFrame({
    'Sales': [10, 20, 30],
    'Inventory': [100, 150, 200]
}, index=['Store_A', 'Store_B', 'Store_C'])

df
```

```python
# Get Sales for Store_B
val = df.loc['Store_B', 'Sales'] # Result: 20
print(val)
```

```python
# Get everything for Store_A and Store_C
subset = df.loc[['Store_A', 'Store_C'], :]
print(subset)
```

```python
# Slicing (Note: In .loc, the end is INCLUDED)
slice_df = df.loc['Store_A':'Store_B', 'Sales'] # Includes Store_B
print(slice_df)
```

### 2. Using .iloc (Integer-based)

Syntax: df.iloc[row_index, column_index]

```python
# Get the first row, first column
val2 = df.iloc[0, 0] # Result: 10
print(val2)
```

```python
# Get the last row (using -1 like a list)
last_row = df.iloc[-1, :]
print(last_row)
```

```python
# Slicing (Note: In .iloc, the end is EXCLUDED, standard Python style)
slice_df2 = df.iloc[0:2, 0:1] # Gets rows 0, 1 and column 0 (excludes 2)
print(slice_df2)
```

### Boolean Indexing

.loc is the standard way to filter rows based on conditions

```python
# Select all stores where Sales are greater than 15
high_sales = df.loc[df['Sales'] > 15, :]
high_sales
```
