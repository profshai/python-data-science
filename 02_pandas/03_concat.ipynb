# Using concat()

```python
import pandas as pd
import numpy as np
```

```python
# combine two Series... default is by rows

s1 = pd.Series(["a", "b"])
s2 = pd.Series(["c", "d"])
pd.concat([s1, s2])
```

```python
# ignoring the original index

pd.concat([s1, s2], ignore_index=True)
```

```python
# Labeling the index

pd.concat([s1, s2], keys=["s1", "s2"], names=["Series name", "Row ID"])
```

```python
# combine two Series by column

s3 = pd.Series(["a", "b"])
s4 = pd.Series(["c", "d"])
df1 = pd.concat([s3, s4], axis = 1)
df1
```

```python
df1 = pd.DataFrame([["a", 1], ["b", 2]], columns=["letter", "number"])
df1
```

```python
df3 = pd.DataFrame(
    [["c", 3, "cat"], ["d", 4, "dog"]], columns=["letter", "number", "animal"])
df3
```

```python
pd.concat([df1, df3], sort=False)
```

```python
# Append a DataFrame at the end of another DataFrame

data1 = {
  "age": [16, 14, 10],
  "qualified": [True, True, True]
}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
  "age": [55, 40],
  "qualified": [True, False]
}
df2 = pd.DataFrame(data2)
df2
```

```python
newdf = pd.concat([df1, df2])
newdf
```
