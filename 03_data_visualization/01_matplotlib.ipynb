# Matplotlib

Bar Chart: df.plot(kind='bar') or df.plot.bar()

Horizontal Bar: df.plot.barh()

Area Plot: df.plot.area()

Box Plot: df.plot.box()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. The Anatomy of a Basic Plot

```python
# 1. Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# 2. Create the plot
plt.plot(x, y, label='Squared Values', color='blue', marker='o')

# 3. Customize
plt.title('My First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)

# 4. Show it
plt.show()
```

## 2. Common Chart Types

### Bar Charts (.bar)

```python
categories = ['Apples', 'Bananas', 'Cherries']
values = [10, 15, 7]

plt.bar(categories, values, color='orange', alpha=0.6)
plt.show()
```

```python
categories = ['Apples', 'Bananas', 'Cherries']
values = [10, 15, 7]

plt.barh(categories, values, color='orange')
plt.show()
```

### Scatter Plots (.scatter)

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y, color='red', s=100) # 's' is size
plt.show()
```

### Histograms (.hist)

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5, edgecolor='black')
plt.show()
```

### Line plots

```python
# Create sample data
df = pd.DataFrame({
    'Sales': [10, 25, 18, 42, 35],
    'Profit': [2, 8, 5, 15, 12]
}, index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])

# Plot all columns
df.plot(title="Monthly Performance", marker='o')
plt.ylabel("Amount ($)")
plt.show()
```

### Pie Chart

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
# Data to plot
sizes = [35, 25, 25, 15]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

# Create the pie chart
plt.pie(sizes, labels=labels, explode=(0.1, 0, 0, 0))
plt.axis('equal')
plt.show()
```

## 3. Subplots (Multiple Charts in One)

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create 1 row, 2 columns of charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# First chart
ax1.plot(x, y)
ax1.set_title('Line Plot')

# Second chart
ax2.scatter(x, y)
ax2.set_title('Scatter Plot')

plt.tight_layout() # Prevents overlapping labels
plt.show()
```

```python
# Create sample data
df = pd.DataFrame({
    'Sales': [10, 25, 18, 42, 35],
    'Profit': [2, 8, 5, 15, 12]
}, index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
df
```

```python
# Plot all columns
df.plot(title="Monthly Performance", marker='o')
plt.ylabel("Amount ($)")
plt.show()
```

```python
# Another plot with the mean line

ax = df.plot(kind='line')
ax.axhline(df['Sales'].mean(), color='red', linestyle='--') # Add a mean line
plt.show()
```
